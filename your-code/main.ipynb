#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
month_avg_steps = fitbit.groupby(['Months_encoded'])['Steps'].median()

fig = plt.figure(figsize= 10,10)
plt.bar(month_avg_steps.index, month_avg_steps.values, color= 'orange')
plt.xticks(month_avg_steps.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.savefig('figures/monthly_steps.png')
plt.show
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND

work_weekend = fitbit.groupby(['Work_or_Weekend'])['Steps'].median()

fig = plt.figure(figsize= (10, 10))
plt.bar(work_weekend.index, work_weekend.values, color= 'green')
plt.xticks(work_weekend.index, ['Work', 'Weekend'])
plt.savefig('figures/week_weekend.png')
plt.show
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
cols = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for col in cols:
    plt.figure()
    plt.title(col + 'vs Steps')
    plt.xlabel('Steps')
    plt.ylabel(col)
    plt.scatter(fitbit[col], fitbit['Steps'])
    plt.savefig(col + '.png')
    plt.show
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
gender = titanic['Gender'].value_counts(normalize=True) * 100
plt.pie(gender, labels=['Male', 'Female'])
plt.show
```

```python
# Method 2 - seaborn

gender = titanic['Gender'].astype('category')
gender = gender.cat.codes
colors = sns.color_palette('bright')[0:2]
plt.pie(gender, labels = ('Male', 'Female'), colors = colors, autopct='%.0f%%')
plt.show()
```
